```python
# create function that replace missing values of columns with text  'missing'

def replace_missing():
    # replace missing values of 'summary' column with text 'missing'
    df['summary'].fillna('missing', inplace=True)
    # replace missing values of 'space' column with text 'missing'
    df['space'].fillna('missing', inplace=True)
    # replace missing values of 'neighborhood_overview' column with text 'missing'
    df['neighborhood_overview'].fillna('missing', inplace=True)
    
    # replace missing values of 'transit' column with text 'missing'
    df['transit'].fillna('missing', inplace=True)
    # replace missing values of 'thumbnail_url' column with text 'missing'
    df['thumbnail_url'].fillna('missing', inplace=True)
    # replace missing values of 'host_since' column with text 'missing'
    df['host_since'].fillna('missing', inplace=True)
    # replace missing values of 'host_location' column with text 'missing'
    df['host_location'].fillna('missing', inplace=True)
    # replace missing values of 'host_about' column with text 'missing'
    df['host_about'].fillna('missing', inplace=True)
    # replace missing values of 'host_response_time' column with text 'missing'
    df['host_response_time'].fillna('missing', inplace=True)
    # replace missing values of 'host_response_rate' column with text 'missing'
    df['host_response_rate'].fillna('missing', inplace=True)
    # replace missing values of 'host_acceptance_rate' column with text 'missing'
    df['host_acceptance_rate'].fillna('missing', inplace=True)
    # replace missing values of 'host_neighbourhood' column with text 'missing'
    df['host_neighbourhood'].fillna('missing', inplace=True)
    # replace missing values of 'host_thumbnail_url' column with text 'missing'
    df['host_thumbnail_url'].fillna('missing', inplace=True)
    # replace missing values of 'host_picture_url' column with text 'missing'
    df['host_picture_url'].fillna('missing', inplace=True)
    # replace missing values of 'host_is_superhost' column with text 'missing'
    df['host_is_superhost'].fillna('missing', inplace=True)
    # replace missing values of 'host_neighbourhood' column with text 'missing'
    df['host_neighbourhood'].fillna('missing', inplace=True)
    # replace missing values of 'bathrooms' column with mode
    df['bathrooms'].fillna(df['bathrooms'].mode(), inplace=True)
    # replace missing values of 'bedrooms' column with  mode
    df['bedrooms'].fillna(df['bedrooms'].mode(), inplace=True)
    # replace missing values of 'beds' column with mode
    df['beds'].fillna(df['beds'].mode(), inplace=True)
    # replace missing values of 'square_feet' column with median
    df['square_feet'].fillna(df['square_feet'].median(), inplace=True)
    # replace missing values of 'weekly_price' column with median
    return df


    
    
```
